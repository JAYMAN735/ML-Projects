```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
categories=["A","B","C","D"]
val1=[4,5,3,6]
val2=[2,5,4,3]

x=np.arange(len(categories))



plt.figure(figsize=(10,6))
plt.bar(x,val1,color="skyblue",width=0.6)
plt.xticks(x,categories)
plt.xlabel("Categories")
plt.title("column chart")
plt.ylabel("Values")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.barh(x,val1,color="orange")
plt.xticks(x,categories)
plt.xlabel("Categories")
plt.title("bar graph")
plt.ylabel("Values")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(x,val1,color="skyblue",label="val1")
plt.bar(x,val2,bottom=val1,color="lightgreen",label="val2")
plt.xticks(x,categories)
plt.xlabel("Categories")
plt.title("stacked bar graph")
plt.ylabel("Values")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.barh(x,val1,height=0.4,color="skyblue",label="val1")
plt.barh(x,val2,height=0.4,color="lightgreen",label="val2")
plt.xticks(x,categories)
plt.xlabel("Categories")
plt.title("stacked bar graph")
plt.ylabel("Values")
plt.legend()
plt.show()
```

```python
x=np.arange(1,6)
y1=np.array([2,4,6,5,7])
y2=np.array([1,3,4,6,5])

plt.figure(figsize=(8,5))
plt.stackplot(x,y1,y2,labels=["series1","series2"],alpha=0.7)
plt.xlabel("X-axis")
plt.ylabel("values")
plt.title("Area Chart")
plt.legend()
plt.show()
```

```python
fig,ax1=plt.subplots(figsize=(8,5))
ax2=ax1.twinx()
ax1.bar(x,y1,color="skyblue",label="bar chart")
ax2.plot(x,y2,color="orange",marker="o",label="line data")

ax1.set_xlabel("X-axis")
ax1.set_ylabel("bar chart")
ax2.set_ylabel("line chart")
plt.title("line values")
plt.show()
```

```python
#mekko chart
widths=[0.5,1.0,1.5,1.0,2.0]
values=[3,5,2,6,4]

lefts=np.cumsum([0]+widths[:-1])
plt.figure(figsize=(8,5))
plt.bar(lefts,values,width=widths,color="lightgreen",edgecolor="black")
plt.title("Mekko chart (simulated)")
plt.xlabel("variable width categories")
plt.ylabel("values")
plt.show()
```
