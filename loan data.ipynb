```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/loan_data.csv')
```

```python
df.head(10)
```

```python
df.isnull().sum()
```

```python
df_filtered = df[df['person_age'] <= 80]
print(f"Original DataFrame shape: {df.shape}")
print(f"Filtered DataFrame shape: {df_filtered.shape}")
df=df_filtered
display(df_filtered.head())
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 7))
sns.countplot(data=df, x='person_home_ownership', hue='loan_status', palette='coolwarm')
plt.title('Person Home Ownership vs. Loan Status')
plt.xlabel('Person Home Ownership')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Loan Status', labels=['No Loan', 'Loan Taken'])
plt.tight_layout()
plt.show()
```

```python
age_counts = df['person_age'].value_counts().sort_index()
display(age_counts)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.barplot(x=age_counts.index, y=age_counts.values, hue=age_counts.index, palette='viridis', legend=False)
plt.title('Number of Persons per Age')
plt.xlabel('Age')
plt.ylabel('Number of Persons')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter the DataFrame for the 60-79 age group
df_60_79 = df[df['age_interval'] == '60-79']

plt.figure(figsize=(10, 6))
sns.countplot(data=df_60_79, x='age_interval', hue='loan_status', palette='coolwarm')
plt.title('Loan Status for Age Group 60-79')
plt.xlabel('Age Interval')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Loan Status', labels=['No Loan', 'Loan Taken'])
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np

# Define age bins with a difference of 20
min_age = int(df['person_age'].min())
max_age = int(df['person_age'].max())

bins = np.arange(min_age, max_age + 20, 20)
labels = [f'{i}-{i+19}' for i in bins[:-1]]

# Handle cases where max_age might fall exactly on a bin boundary
if max_age > bins[-2]:
    labels.append(f'{bins[-2]}-{max_age}')

df['age_interval'] = pd.cut(df['person_age'], bins=bins, labels=labels[:len(bins)-1], right=False)

age_interval_counts = df['age_interval'].value_counts().sort_index()
display(age_interval_counts)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(x=age_interval_counts.index, y=age_interval_counts.values, hue=age_interval_counts.index, palette='viridis', legend=False)
plt.title('Number of Persons per Age Interval (Difference of 20)')
plt.xlabel('Age Interval')
plt.ylabel('Number of Persons')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
