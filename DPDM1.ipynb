```python
df['Journey_Day'] = pd.to_datetime(df['Date_of_Journey'], format='%d/%m/%Y').dt.day
df['Journey_Month'] = pd.to_datetime(df['Date_of_Journey'], format='%d/%m/%Y').dt.month
df['Journey_Year'] = pd.to_datetime(df['Date_of_Journey'], format='%d/%m/%Y').dt.year
```

```python
df.drop(['Date_of_Journey'], axis=1, inplace=True)
```

```python
display(df.head())
```

```python
df.info()
```

```python
most_common_route = df['Route'].value_counts().index[0]
most_common_route_count = df['Route'].value_counts().iloc[0]
print(f"The most common route is '{most_common_route}' with a count of {most_common_route_count}.")
```

```python
df.dropna(subset=['Total_Stops'], inplace=True)
print("Rows with NaN values in 'Total_Stops' column have been deleted.")

display(df.head())
display(df.info())
```
