```python
import pandas as pd
import sqlite3
```

### Loading `website_sales.db` (SQLite Database)

First, I will connect to the SQLite database and list the tables to identify the relevant sales table. Then, I will load the data from that table into a DataFrame.

```python
conn = sqlite3.connect('/content/website_sales.db')
cursor = conn.cursor()

# Get table names from the database
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = cursor.fetchall()
print("Tables in website_sales.db:", tables)

# Assuming 'sales' is the table name, load it into a DataFrame
# You might need to adjust 'sales' if the table name is different
website_sales_df = pd.read_sql_query("SELECT * FROM sales", conn)
conn.close()

display(website_sales_df.head())
```

### Loading `amazon_sales.csv`

Next, I will load the `amazon_sales.csv` file into a pandas DataFrame.

```python
amazon_sales_df = pd.read_csv('/content/amazon_sales.csv')
display(amazon_sales_df.head())
```

### Loading `flipkart_sales.json`

Finally, I will load the `flipkart_sales.json` file into a pandas DataFrame.

```python
flipkart_sales_df = pd.read_json('/content/flipkart_sales.json')
display(flipkart_sales_df.head())
```

```python
# Convert the 'date' column in amazon_sales_df to datetime objects first
amazon_sales_df['date'] = pd.to_datetime(amazon_sales_df['date'], format='%d-%m-%Y')

# Now, format the datetime objects to 'YYYY-MM-DD' string format
amazon_sales_df['date'] = amazon_sales_df['date'].dt.strftime('%Y-%m-%d')

print("Date format in amazon_sales_df updated to 'YYYY-MM-DD':")
display(amazon_sales_df.head())
```

```python
amazon_sales_df = amazon_sales_df.rename(columns={'units_sold': 'units'})
print("Column 'units_sold' renamed to 'units' in amazon_sales_df:")
display(amazon_sales_df.head())
```

```python
flipkart_sales_df = flipkart_sales_df.rename(columns={'product_name': 'product', 'quantity': 'units', 'unit_price': 'price'})
print("Columns renamed in flipkart_sales_df:")
display(flipkart_sales_df.head())
```

```python
print("--- Info for website_sales_df ---")
print(website_sales_df.info())
display(website_sales_df.head())
```

### `amazon_sales_df` (Amazon Sales)
This DataFrame contains sales data from Amazon. The 'date' column has been formatted to 'YYYY-MM-DD' and 'units_sold' has been renamed to 'units'.


```python
print("\n--- Info for amazon_sales_df ---")
print(amazon_sales_df.info())
display(amazon_sales_df.head())
```

### `flipkart_sales_df` (Flipkart Sales)
This DataFrame contains sales data from Flipkart. The columns 'product_name', 'quantity', and 'unit_price' have been renamed to 'product', 'units', and 'price' respectively.


```python
print("\n--- Info for flipkart_sales_df ---")
print(flipkart_sales_df.info())
display(flipkart_sales_df.head())
```

```python
flipkart_sales_df['date'] = flipkart_sales_df['date'].dt.strftime('%Y-%m-%d')
print("Date format in flipkart_sales_df updated to 'YYYY-MM-DD' and type converted to object:")
display(flipkart_sales_df.head())
```

```python
print(flipkart_sales_df.info())
```

```python
# Add a 'source' column to each DataFrame
amazon_sales_df['source'] = 'amazon'
flipkart_sales_df['source'] = 'flipkart'

# Merge the two DataFrames using concat
combined_sales_df = pd.concat([amazon_sales_df, flipkart_sales_df], ignore_index=True)

print("Combined DataFrame head:")
display(combined_sales_df.head())

print("\nCombined DataFrame info:")
print(combined_sales_df.info())
```

```python
combined_sales_df['revenue'] = combined_sales_df['units'] * combined_sales_df['price']
print("Combined DataFrame with 'revenue' column:")
display(combined_sales_df.head())
```

```python
website_sales_df['source'] = 'website'
print("website_sales_df with 'source' column:")
display(website_sales_df.head())
```

```python
website_sales_df['price'] = website_sales_df['revenue'] / website_sales_df['units']
print("website_sales_df with new 'price' column:")
display(website_sales_df.head())
```

```python
final_combined_sales_df = pd.concat([combined_sales_df, website_sales_df], ignore_index=True)

print("Final Combined DataFrame head:")
display(final_combined_sales_df.head())

print("\nFinal Combined DataFrame info:")
print(final_combined_sales_df.info())
```
